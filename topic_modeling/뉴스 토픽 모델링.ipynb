```python
data = pd.concat([data1, data2, data3, data4, data5, data6, data7, data8], axis=0)
data
```

```python
df = data.dropna()
df = df.sort_values(by=['date'], axis=0)
df = df.reset_index().drop(["index"], axis=1)
df = df.drop(["Unnamed: 0"], axis=1)
df
```

```python
from nltk.tokenize import sent_tokenize
df['Sentence_tokenize'] = df.text.apply(lambda x : sent_tokenize(x))
df.head()
```

```python
from konlpy.tag import Mecab
mecab = Mecab()

tokenized_docs = []
for text in df.Sentence_tokenize:
    tmp = []
    for sentence in text:
        tmp.append([word for word in mecab.nouns(sentence) if len(word)>1])
    tokenized_docs.append(tmp)
```

```python
df['Word_tokenize'] = tokenized_docs
```

# LDA Mallet

```python
import warnings
warnings.filterwarnings("ignore")

import os
from gensim import corpora
from gensim.models.wrappers import LdaMallet

os.environ['Mallet_HOME'] = 'C:\\Mallet'
mallet_path = 'C:\\Mallet\\bin\\mallet' 

topic_keywords = []

for tokenized_sentence in df.Word_tokenize[:100]:
    id2word=corpora.Dictionary(tokenized_sentence)
    texts = tokenized_sentence
    corpus=[id2word.doc2bow(text) for text in texts]
    
    ldamallet = LdaMallet(mallet_path, corpus=corpus, num_topics=1, id2word=id2word)
    topic_words_list = ldamallet.show_topic(0)
    topic_keywords.append([word for word, prop in topic_words_list])
```

```python
topic_keywords
```

```python
df_final = pd.DataFrame({'Date':df.date[:100], 'Topic_keywords':topic_keywords, 'Text':df.text[:100]})
df_final
```
