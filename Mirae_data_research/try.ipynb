```python
start_date = 201907
end_date = 202008
start_date = datetime.strptime(str(start_date),'%Y%m')
end_date = datetime.strptime(str(end_date),'%Y%m')

while start_date <= end_date:
    last_day = start_date + relativedelta(months=+1)
    
    print(start_date)
    print(last_day)
    
    # 1개월 기간 추출
    this_time = traded_stocks_info[(traded_stocks_info['기준일자']>=start_date) & (traded_stocks_info['기준일자']<last_day)]
    
    # 필요항목 계산
    ## 종목코드들
    codes = this_time['종목번호'].tolist()
#     copied[['test1', 'test2', 'test3', 'test4', 'test5']] = codes.apply(lambda x: get_std(x, this_time))
    
    col_start_date = start_date.strftime("%Y%m")
    
    for code in tqdm_notebook(codes):
        stop_trading_duration, std_year, change_rate, cal_max, cal_min = get_std(code, this_time)
        copied.at[code, '{}_정지이력'.format(col_start_date)] = stop_trading_duration
        copied.at[code, '{}_변동성'.format(col_start_date)] = std_year
        copied.at[code, '{}_모멘텀'.format(col_start_date)] = change_rate
        copied.at[code, '{}_최고상승률'.format(col_start_date)] = cal_max
        copied.at[code, '{}_최고하락율'.format(col_start_date)] = cal_min
        
    
    
    
    

    
    start_date = last_day
    print(start_date)
```

```python
def get_std(row, this_time):
#     print(row)
    code = row
    
    # 거래정지이력
    stop_trading_duration = len(this_time[(this_time['종목번호'] == code) & (this_time['거래량'] == 0)])
    
    cal_data = this_time[(this_time['종목번호'] == code) & (this_time['거래량'] > 0)]['종목종가']
    avg = cal_data.mean()
    std = cal_data.std()
    
    # 표준편차
    try :
        std_year = ((std/avg)/np.sqrt(len(cal_data)))*np.sqrt(252) * 100
    except:
        std_year = None
        
    # 1개월 가격변화율
    try:
        change_rate = np.log(cal_data.tolist()[-1]/cal_data.tolist()[0])
    except:
        change_rate = None
    
    # 최고 상승 & 최고 하락
    try:
        cal_max = np.log(cal_data.max()/cal_data.tolist()[0])
    except:
        cal_max = None
    try:
        cal_min = np.log(cal_data.min()/cal_data.tolist()[0])
    except:
        cal_min = None
    
    
    return stop_trading_duration, std_year, change_rate, cal_max, cal_min
```
